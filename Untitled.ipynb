### EDA

```python
def preprocess_duration(x):
    if "h" not in x:
        x = '0h' + ' ' +x
    elif "m" not in x:
        x = x + ' ' + '0m'

    return x
```

```python
train_df['Duration'] = train_df['Duration'].apply(preprocess_duration)
```

```python
train_df['Duration_hour'] = pd.to_timedelta(train_df['Duration']).dt.components.hours
train_df['Duration_minutes'] = pd.to_timedelta(train_df['Duration']).dt.components.minutes
```

```python
train_df.head(2)
```

```python
cols_to_drop = ['Dep_Time','Arrival_Time','Duration','Date_of_Journey']

train_df.drop(cols_to_drop, axis=1, inplace=True)
```

```python
train_df.columns
```

```python
train_df["Duration_total_mins"] = train_df["Duration_hour"] * 60 + train_df["Duration_minutes"]
```

```python
def plot(df,col):
    fig, (ax1,ax2,ax3) = plt.subplots(3,1)
    sns.distplot(df[col],ax=ax1)
    sns.boxplot(data=df,x=col,ax=ax2)
    sns.distplot(df[col],ax=ax3,kde=False)

plot(train_df,"Price")
```

```python
fig, axes = plt.subplots(1, 2, figsize=(18, 5))

sns.boxplot(x='Airline',   y='Price', data=train_df, ax=axes[0])
sns.boxplot(x='Total_Stops', y='Price', data=train_df, ax=axes[1])

axes[0].set_title('Price by Airline')
axes[1].set_title('Price by Stops')

for ax in axes:
    ax.tick_params(axis='x', rotation=0)

plt.tight_layout()
plt.show()
```

```python
##Removing outliers using IQR 

Q1 = train_df['Price'].quantile(0.25)
Q3 = train_df['Price'].quantile(0.75)
IQR = Q3 - Q1

lower = Q1 - 1.5 * IQR
upper = Q3 + 1.5 * IQR

print(f"Outlier range: below {lower:.0f} or above {upper:.0f}")
print(f"Outliers found: {((train_df['Price'] < lower) | (train_df['Price'] > upper)).sum()}")

# Remove outliers
train_df = train_df[(train_df['Price'] >= lower) & (train_df['Price'] <= upper)]
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Only numeric columns can be correlated
corr_matrix = train_df.corr()

plt.figure(figsize=(12, 8))
sns.heatmap(
    corr_matrix,
    annot=True,          # show numbers inside cells
    fmt=".2f",           # 2 decimal places
    cmap="coolwarm",     # blue = positive, red = negative
    center=0,            # white at 0
    linewidths=0.5,
    linecolor='white',
    square=True
)
plt.title("Correlation Heatmap")
plt.tight_layout()
plt.show()
```

```python
##Checking for Corrlation 

# See which features correlate most with Price
price_corr = train_df.corr()['Price'].drop('Price').sort_values(ascending=False)
print(price_corr)

# Bar chart of correlations with Price
plt.figure(figsize=(10, 5))
price_corr.plot(kind='bar', color=['#2196F3' if x > 0 else '#E53935' for x in price_corr])
plt.axhline(0, color='black', linewidth=0.8)
plt.title('Feature Correlation with Price')
plt.ylabel('Correlation coefficient')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.25,random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from xgboost import XGBRegressor
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error
import numpy as np

models = {
    "Linear Regression":      LinearRegression(),
    "Decision Tree":          DecisionTreeRegressor(random_state=42),
    "Random Forest":          RandomForestRegressor(n_estimators=100, random_state=42),
    "XGBoost":                XGBRegressor(n_estimators=100, random_state=42),
}

results = {}

for name, model in models.items():
    # Train
    model.fit(X_train, y_train)

    # Predict
    y_pred = model.predict(X_test)

    # Evaluate
    r2  = r2_score(y_test, y_pred)
    mae = mean_absolute_error(y_test, y_pred)
    rmse = np.sqrt(mean_squared_error(y_test, y_pred))

    results[name] = {"R2": round(r2, 4), "MAE": round(mae, 2), "RMSE": round(rmse, 2)}
    print(f"{name:25s} | R²: {r2:.4f} | MAE: {mae:.0f} | RMSE: {rmse:.0f}")

```
