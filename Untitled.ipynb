### EDA

```python
def preprocess_duration(x):
    if "h" not in x:
        x = '0h' + ' ' +x
    elif "m" not in x:
        x = x + ' ' + '0m'

    return x
```

```python
train_df['Duration'] = train_df['Duration'].apply(preprocess_duration)
```

```python
train_df['Duration_hour'] = pd.to_timedelta(train_df['Duration']).dt.components.hours
train_df['Duration_minutes'] = pd.to_timedelta(train_df['Duration']).dt.components.minutes
```

```python
cols_to_drop = ['Dep_Time','Arrival_Time','Duration','Date_of_Journey']

train_df.drop(cols_to_drop, axis=1, inplace=True)
```

```python
##Removing outliers using IQR 

Q1 = train_df['Price'].quantile(0.25)
Q3 = train_df['Price'].quantile(0.75)
IQR = Q3 - Q1

lower = Q1 - 1.5 * IQR
upper = Q3 + 1.5 * IQR

print(f"Outlier range: below {lower:.0f} or above {upper:.0f}")
print(f"Outliers found: {((train_df['Price'] < lower) | (train_df['Price'] > upper)).sum()}")

# Remove outliers
train_df = train_df[(train_df['Price'] >= lower) & (train_df['Price'] <= upper)]
```

```python
from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.25,random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from xgboost import XGBRegressor
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error,mean_absolute_percentage_error
import numpy as np

models = {
    "Linear Regression":      LinearRegression(),
    "Decision Tree":          DecisionTreeRegressor(random_state=42),
    "Random Forest":          RandomForestRegressor(n_estimators=100, random_state=42),
    "XGBoost":                XGBRegressor(n_estimators=100, random_state=42),
}

results = {}

for name, model in models.items():
    # Train
    model.fit(X_train, y_train)

    # Predict
    y_pred = model.predict(X_test)

    # Evaluate
    r2  = r2_score(y_test, y_pred)
    mae = mean_absolute_error(y_test, y_pred)
    rmse = np.sqrt(mean_squared_error(y_test, y_pred))
    mape = mean_absolute_percentage_error(y_test, y_pred) * 100


    results[name] = {"R2": round(r2, 4), "MAE": round(mae, 2), "RMSE": round(rmse, 2)}
    print(f"{name:25s} | R²: {r2:.4f} | MAE: {mae:.0f} | RMSE: {rmse:.0f} | MAPE: {mape:.2f}%")

```

```python
from sklearn.model_selection import train_test_split, RandomizedSearchCV, cross_val_score
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error

```

```python
from xgboost import XGBRegressor
from sklearn.model_selection import cross_val_score, RandomizedSearchCV
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error
import numpy as np

xgb = XGBRegressor(random_state=42, n_jobs=-1, verbosity=0)
cv  = cross_val_score(xgb, X_train, y_train, cv=5, scoring='r2')
print(f"Baseline CV R²: {cv.mean():.4f} ± {cv.std():.4f}")
```

```python
params = {
    'n_estimators':     [100, 300, 500, 700],
    'learning_rate':    [0.01, 0.05, 0.1, 0.2],
    'max_depth':        [3, 5, 6, 8],
    'min_child_weight': [1, 3, 5],
    'subsample':        [0.6, 0.8, 1.0],
    'colsample_bytree': [0.6, 0.8, 1.0],
    'reg_alpha':        [0, 0.1, 0.5],
    'reg_lambda':       [1, 1.5, 2],
    'gamma':            [0, 0.1, 0.5],
}

rs = RandomizedSearchCV(
    XGBRegressor(random_state=42, n_jobs=-1, verbosity=0),
    params, n_iter=100, cv=5, scoring='r2',
    n_jobs=1, random_state=42, verbose=1
)
rs.fit(X_train, y_train)

print(f"Best CV R² : {rs.best_score_:.4f}")
print(f"Best Params: {rs.best_params_}")
```

```python
# Remove n_estimators from best_params before unpacking
best_params = {k: v for k, v in rs.best_params_.items() if k != 'n_estimators'}

best_model = XGBRegressor(
    **best_params,
    n_estimators=2000,          # now no conflict
    early_stopping_rounds=50,
    random_state=42, n_jobs=-1, verbosity=0
)

best_model.fit(
    X_train, y_train,
    eval_set=[(X_test, y_test)],
    verbose=False
)

print(f"Best iteration: {best_model.best_iteration}")
```

```python
y_pred = best_model.predict(X_test)

print(f"R²   : {r2_score(y_test, y_pred):.4f}")
print(f"MAE  : ₹{mean_absolute_error(y_test, y_pred):.0f}")
print(f"RMSE : ₹{np.sqrt(mean_squared_error(y_test, y_pred)):.0f}")
print(f"MAPE : {np.mean(np.abs((y_test - y_pred) / y_test)) * 100:.2f}%")

best_model.save_model('xgb_tuned.json')
print("Model saved!")
```

```python
import matplotlib.pyplot as plt

rmse_train = best_model.evals_result()['validation_0']['rmse']

plt.figure(figsize=(8, 4))
plt.plot(rmse_train, color='steelblue', label='Test RMSE')
plt.axvline(best_model.best_iteration, color='green',
            linestyle='--', label=f'Best: {best_model.best_iteration}')
plt.xlabel('Boosting Round')
plt.ylabel('RMSE')
plt.legend()
plt.tight_layout()
plt.show()
```
